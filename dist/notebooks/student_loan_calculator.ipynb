## Helpful Formulas

### Simple monthly payment calculation

$$M = P \cdot \frac{r(1+r)^n}{(1+r)^n - 1}$$

Where:
- $M$ = Monthly payment
- $P$ = Principal loan amount
- $r$ = Monthly interest rate (annual rate รท 12)
- $n$ = Total number of payments (months)
