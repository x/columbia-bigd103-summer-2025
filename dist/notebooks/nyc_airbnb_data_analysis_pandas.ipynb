## Understanding the Data

The dataset contains information about Airbnb listings in NYC. Here are all the columns:

- **id**: Unique identifier for each listing
- **name**: Title/description of the listing (e.g., "Cozy Brooklyn Apartment")
- **host_id**: Unique identifier for the host
- **host_name**: Name of the person hosting
- **neighbourhood_group**: Borough (Manhattan, Brooklyn, Queens, Bronx, Staten Island)
- **neighbourhood**: Specific neighborhood name (e.g., "Williamsburg", "Upper East Side")
- **latitude**: Geographic coordinate (north-south position)
- **longitude**: Geographic coordinate (east-west position)
- **room_type**: Type of accommodation ("Entire home/apt", "Private room", "Shared room", "Hotel room")
- **price**: Price per night in dollars
- **minimum_nights**: Minimum stay requirement
- **number_of_reviews**: Total number of reviews the listing has received
- **last_review**: Date of most recent review
- **reviews_per_month**: Average number of reviews per month
- **calculated_host_listings_count**: How many total listings this host has
- **availability_365**: Days available to book per year
- **number_of_reviews_ltm**: Number of reviews in the last 12 months
- **license**: Business license information (mostly empty)

## Hints

- Use `df.groupby('column')['price'].mean()` for simple grouping
- Use `df.groupby('column')['price'].agg(['mean', 'median', 'count'])` for multiple stats
- Use `df.groupby(['col1', 'col2'])` for multi-level grouping
- Remember that groupby operations return Series or DataFrames that can be further manipulated
- Use `.reset_index()` to convert a grouped result back to a regular DataFrame
- Boolean indexing: `df[df['column'] > value]` filters rows
- Create new columns with: `df['new_col'] = df['old_col'].apply(some_function)`

## Extra Pandas Tips

- `df.describe()` gives you quick statistics
- `df.value_counts()` counts unique values
- `df.sort_values()` sorts by column(s)
- `df.merge()` joins dataframes
- `df.pivot_table()` creates summary tables
- `df.corr()` shows correlations between numeric columns
