## Exercise: Grade Calculator

Create a program that takes a numeric test score and provides detailed feedback.

### Your Task

1. Ask the user for their test score (0-100)
2. Determine their letter grade using these ranges:
   - 90-100: A
   - 80-89: B
   - 70-79: C
   - 60-69: D
   - Below 60: F
3. For each grade, also print:
   - A custom message about their performance
   - How many points they need for the next grade level (except for A)
   - If they got an A with 100 points, print a special "Perfect Score!" message

### Example Output

```
Enter your test score: 82
Your grade: B
Good job! You're doing well.
You need 8 more points for an A.
```

```
Enter your test score: 100
Your grade: A
Excellent work!
Perfect Score! Outstanding!
```

```
Enter your test score: 58
Your grade: F
Don't give up! See me for extra help.
You need 2 more points for a D.
```

### Tips
- Use `elif` statements for each grade range
- Calculate `points_needed` by subtracting current score from next threshold
- Remember to handle the special case of a perfect 100
- Test with scores like: 100, 92, 85, 77, 65, 58

### Bonus Challenges
1. Add input validation: make sure the score is between 0 and 100
2. Add plus/minus grades (A+ for 97-100, A for 93-96, A- for 90-92, etc.)

````python
# Get the score
score = int(input("Enter your test score: "))

# Determine grade and feedback
if score >= 90:
    grade = "A"
    message = "Excellent work!"


print(f"Your score: {score}/100")
print(f"Your grade: {grade}")
print(message)
````
