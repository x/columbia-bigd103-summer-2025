# Quick Intro to Google Colab

In this first assignment, you'll become familiar with the Colab Notebook.

**This is a Colab Notebook.**

Under the hood, "Colab Notebooks" are a special Google-hosted version of what we call "Jupyter Notebooks." These notebooks mix text, code, and data visualization in one document.

## Structure
- **Filename** (top-left corner) ending in `.ipynb`
- **Path** in Google Drive (which acts like a file system)
- **Cells** - this is a text cell; you can also have code cells

## Code Cells
- Code cells contain Python code and can be executed with `Shift+Enter` or by clicking the ▶️ button
- If you use `print(...)`, the output appears below the cell
- The last line of a cell automatically displays its result (this is REPL behavior)

*This might seem like a lot, but you'll be spending lots of time here over the next three weeks, so get comfortable!*
