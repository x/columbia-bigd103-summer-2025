## Exercise: The Guessing Game

Create a game where the computer picks a random number and the player tries to guess it!

### Your Task

Build a guessing game with these features:
1. Computer picks a random number between 1 and 100
2. Player gets 10 attempts to guess
3. After each wrong guess, tell them if the secret number is higher or lower
4. If they guess correctly, they win!
5. If they run out of guesses, they lose

### Example Game Play

```
Welcome to the Number Guessing Game!
I'm thinking of a number between 1 and 100.
You have 10 guesses.

Guess 1: 50
Too low! The secret number is higher.

Guess 2: 75
Too high! The secret number is lower.

Guess 3: 62
Too low! The secret number is higher.

Guess 4: 68
Congratulations! You got it in 4 guesses!
```

### Tips
- Use `break` to exit the loop when they guess correctly
- The `else` in `if/elif/else` handles the "too high" case
- Consider adding a `print` statement to "whisper" the secret number to you as you're working out the kinks in your game.

### Bonus Challenges
1. Tell them how many guesses they have left after each attempt
2. Ask the user for a difficulty level (easy: 1-50, medium: 1-100, hard: 1-200)
3. Add a play again feature using a while loop and keep track of their lowest number of guesses.




````python
# Import random number generator
from random import randint

# Choose Difficulty
while True:
    difficulty = input("What is your difficulty level (easy, medium, hard)? ")
    if difficulty == "easy":
        max_randint = 50
        break
    elif difficulty == "medium":
        max_randint = 100
        break
    elif difficulty == "hard":
        max_randint = 200
        break
    else:
        print(f"Invalid difficulty '{difficulty}'")

max_guesses = 10
best_score = max_guesses + 1

while True:

    # Generate secret number
    secret_number = randint(1, max_randint)

    # Game loop
    for guess_count in range(1, max_guesses + 1):
        guess = int(input(f"Guess {guess_count}: "))

        # Hints
        if guess < secret_number:
            print("WRONG! The secret number is higher.")
        elif guess > secret_number:
            print("WRONG! The secret number is lower.")

        # Win Condition
        else:
            print(f"CORRECT! The scret number was {guess}.")
            print(f"Score: {guess_count}")
            if guess_count < best_score:
                best_score = guess_count
                print(f"New Best Score: {best_score}")
            break

        # Guesses Remaining
        guesses_left = max_guesses - guess_count
        print(f"You have {guesses_left} guesses remaining")

    continue_game = input("Would you like to play again (yes, no)? ")
    if continue_game == "no":
        print(f"Your best score was: {best_score}")
        break
````
